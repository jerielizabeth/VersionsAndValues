Note: So we are getting individuals by the individual repo, so if the individual repo is not tagged as "DH", then we are not seeing it. The center at Princeton is interesting, in that there is a corporate account that is responsible for the pushes, which obscures which individuals are participating.

So, how can we shift this? 

1. Has a user contributed to a DH flagged repository? If yes, what other repositories have they contributed to? 



```python
import requests
from tqdm import tqdm
import apikey


auth_token = apikey.load("DH_GITHUB_DATA_PERSONAL_TOKEN")

auth_headers = {'Authorization': f'token {auth_token}','User-Agent': 'request'}
```

```python
def get_search_api_data(query):
    # Thanks https://stackoverflow.com/questions/33878019/how-to-get-data-from-all-pages-in-github-api-with-python 
    try:
        response = requests.get(f"{query}", headers=auth_headers)
        response_data = response.json()

        while "next" in response.links.keys():
            url = response.links['next']['url']
            response = requests.get(url, headers=auth_headers)
            response_data.extend(response.json())
    except:
        print(f"Error with URL: {url}")

    return response_data
```

```python
get_api_data("https://api.github.com/users/jerielizabeth")
```

```python
"""
From user data- 

login:
id:
url:
name:
company
blog:
location
bio:
public_repos:
public_gists:
followers:
following:
total_private_repos:
collaborators:

followers_url
subscriptions_url
organizations_url
repos_url
"""
```

```python
contributor_data = []

fields = ['login', 'id', 'url', 'name', 'company','blog', 'location', 'bio', 'public_repos', 'public_gists', 'followers', 'following', 'total_private_repos', 'collaborators', 'followers_url', 'subscription_url', 'organizations_url', 'repos_url']

for user_url in contributors_df['url'].tolist():
    # print(user_url)
    try:
        user_data = get_api_data(user_url)
        
        filtered_data = {}
        for field in fields:
            try:
                filtered_data[field] = user_data[field]
            except:
                filtered_data[field] = 'NaN'

        contributor_data.append(filtered_data)
    except:
        print(f"{user_url} failed")
```

```python
expanded_contributors_df = pd.DataFrame.from_dict(contributor_data)
expanded_contributors_df
```

```python
expanded_contributors_df = expanded_contributors_df.drop_duplicates()
```

```python
expanded_contributors_df.to_csv("../data/repos_topic_dh_contributors_expanded.csv", index=False)
```
